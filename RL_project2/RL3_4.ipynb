## Data based Reward matrix 생성

```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
np.set_printoptions(threshold=np.inf)

def action(state, Q, epsilon, pi_0, N, channel):
    action_direction = np.arange(N)
    if np.random.rand() < epsilon: # 확률 e로 무작위 행동을 선택함
        action = random.choice(channel)
    else: # Q가 알려주는 방향으로 행동을 선택함
        action = action_direction[np.nanargmax(Q[state, :])]
    return action - 1

def next_state(state, action, N): # 다음 상태가 무엇인가
    action_direction = np.arange(N)
    s_next = action_direction[action]  # 현재 action에 따른 다음 상태
    return s_next
    
### 정책 파라미터 theta_0을 무작위 행동 정책 pi로 변환하는 함수
def simple_convert_into_pi_from_theta(theta):
    [m, n] = theta.shape
    pi = np.zeros((m, n))
    for i in range(0, m):
        pi[i, :] = theta[i, :] / np.nansum(theta[i, :])
    pi = np.nan_to_num(pi)
    return pi
    
### Q러닝 알고리즘으로 행동가치 함수 Q를 수정
def Q_learning(state, action, s_next, reward, Q, eta, gamma):
    Q[state, action] = Q[state, action] + eta * (reward + gamma*np.nanmax(Q[s_next,:]) - Q[state, action])
    return Q

def make_reward_data(N, T, channel):
    data = []
    reward_matrix = np.zeros((N,T))
    
    for t in range(T):
        data.append(random.choice(channel))
        label = 0
        
        with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            line_count = 0

            for row in csv_reader:
                if line_count is 0:
                    line_count = 1
                else:
                    if int(row[0]) == data[t]:
                        label = str(row[2])
        
        with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            line_count = 0

            for row in csv_reader:
                if line_count is 0:
                    line_count = 1
                else:
                    if int(row[0]) == data[t]:
                        try:
                            weight = field[str(row[2])]
                            reward_matrix[data[t]-1,t] += 1 * weight
                        except KeyError:
                            reward_matrix[data[t]-1,t] += 0
                    else:
                        if str(row[2]) == label:
                            try:
                                weight = field[str(row[2])]
                                reward_matrix[int(row[0])-1,t] += 0.1 * weight
                            except KeyError:
                                reward_matrix[int(row[0])-1,t] += 0
    return reward_matrix

reward_matrix = make_reward_data(274, 1000, channel)
print(reward_matrix[1, :])
```

## User TV tendency prediction

```python
### 테스트 알고리즘
def test_algorithm(reward_matrix, max_epsilon, min_epsilon, eta, gamma, channel, time, threshold):
    N = reward_matrix.shape[0]
    cumulative_rewards = np.zeros(N)
    theta_0 = np.ones((N,N)) #initialize 정책 파라미터 (가능한 정책 모두를 표기)
    pi_0 = simple_convert_into_pi_from_theta(theta_0) #initialize 행동 정책 (가능한 정책을 확률로 표현)
    [a, b] = theta_0.shape
    Q = np.random.rand(a, b) * theta_0 * 0.5
    #Q = np.zeros((N,N))
    channels = []
    state = random.choice(channel) - 1
    decreased_epsilon = max_epsilon
    
    for t in range(time):
        #print(state+1)
        chosen_ch = action(state, Q, decreased_epsilon, pi_0, N, channel)
        reward = sum(reward_matrix[state, :] * 0.1)
        #print(reward)
        s_next = chosen_ch
        Q = Q_learning(state, chosen_ch, s_next, reward, Q, eta, gamma)
        state = s_next
        
        if max_epsilon >= min_epsilon:
            decreased_epsilon = decreased_epsilon - (max_epsilon - min_epsilon)/time # epsilon 점차적으로 감소
        else:
            print("Max epsilon should bigger than min epsilon")
            raise RuntimeError
        
        if state in channel:
            #print("state", state)
            cumulative_rewards[state] += reward
        
    for ch,value in enumerate(cumulative_rewards):
        if value > threshold:
            channels.append(ch)
            
    return channels


N = 274
data = 10000
time = 20

max_epsilon = 0.01
min_epsilon = 0.01
eta = 0.1
gamma = 0.1
threshold = 0.0 #5.0

reward_matrix = make_reward_data(N,data,channel)
#print("실제로 방송 본 채널\n")
#print(skip_channel)
#print()
random.seed(0)
# 학습 알고리즘 테스트.
results = test_algorithm(reward_matrix, max_epsilon, min_epsilon, eta, gamma, channel, time, threshold)
for i,value in enumerate(results):
    if value not in channel:
        print(value)
        results.remove(value)
print("\n이 채널들 빼고 다 스킵:\n", results)

labels = []
for i in results:
    with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0

        for row in csv_reader:
            if line_count is 0:
                line_count = 1
            else:
                if int(row[0]) == int(i):
                    labels.append(row[2])
print(labels)
```
