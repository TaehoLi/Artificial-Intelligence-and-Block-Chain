## Data based Reward matrix

```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
np.set_printoptions(threshold=np.inf)

def action(state, Q, epsilon, pi_0, N, channel):
    random.seed(567)
    np.random.seed(567)
    action_direction = np.arange(N)
    if np.random.rand() < epsilon: # 확률 e로 무작위 행동을 선택함
        action = random.choice(channel)
    else: # Q가 알려주는 방향으로 행동을 선택함
        action = action_direction[np.nanargmax(Q[state, :])]
    return action

def next_state(state, action, N): # 다음 상태가 무엇인가
    action_direction = np.arange(N)
    s_next = action_direction[action]  # 현재 action에 따른 다음 상태
    return s_next
    
### 정책 파라미터 theta_0을 무작위 행동 정책 pi로 변환하는 함수
def simple_convert_into_pi_from_theta(theta):
    [m, n] = theta.shape
    pi = np.zeros((m, n))
    for i in range(0, m):
        pi[i, :] = theta[i, :] / np.nansum(theta[i, :])
    pi = np.nan_to_num(pi)
    return pi
    
### Q러닝 알고리즘으로 행동가치 함수 Q를 수정
def Q_learning(state, action, s_next, reward, Q, eta, gamma):
    Q[state, action] = Q[state, action] + eta * (reward + gamma*np.nanmax(Q[s_next,:]) - Q[state, action])
    return Q

def make_reward_data(N, T, channel):
    data = []
    reward_matrix = np.zeros((N,T))
    random.seed(82)
    for t in range(T):
        data.append(random.choice(channel))
        label = 0
        
        with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            line_count = 0

            for row in csv_reader:
                if line_count == 0:
                    line_count = 1
                else:
                    if int(row[0]) == (data[t]+1):
                        label = str(row[2])
        
        with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            line_count = 0

            for row in csv_reader:
                if line_count == 0:
                    line_count = 1
                else:
                    if int(row[0]) == (data[t]+1):
                        weight = field[str(row[2])]
                        reward_matrix[data[t],t] += 1 * weight
                    else:
                        if str(row[2]) == label:
                            weight = field[str(row[2])]
                            reward_matrix[int(row[0])-1,t] += 0.1 * weight
    return reward_matrix

reward_matrix = make_reward_data(274, 1000, channel)
print(reward_matrix[1, :])
```

## Channel recommendation

```python
from pprint import pprint

### 테스트 알고리즘
def test_algorithm(reward_matrix, max_epsilon, min_epsilon, eta, gamma, channel, time, threshold):
    N = reward_matrix.shape[0]
    cumulative_rewards = np.zeros(N)
    theta_0 = np.ones((N,N)) #initialize 정책 파라미터 (가능한 정책 모두를 표기)
    pi_0 = simple_convert_into_pi_from_theta(theta_0) #initialize 행동 정책 (가능한 정책을 확률로 표현)
    [a, b] = theta_0.shape
    Q = np.random.rand(a, b) * theta_0 * 0.5
    #Q = np.zeros((N,N))
    channels = []
    state = random.choice(channel)
    decreased_epsilon = max_epsilon
    reward_matrix = np.array(reward_matrix)
    
    for t in range(time):
        chosen_ch = action(state, Q, decreased_epsilon, pi_0, N, channel)
        reward = sum(reward_matrix[state, :] * 0.1)
        s_next = chosen_ch
        Q = Q_learning(state, chosen_ch, s_next, reward, Q, eta, gamma)
        
        stack = np.zeros(N)
        with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            line_count = 0

            for row in csv_reader:
                if line_count is 0:
                    line_count = 1
                else:
                    if int(row[0]) == state+1:
                        label = str(row[2])
        
        with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
            csv_reader = csv.reader(csv_file, delimiter=',')
            line_count = 0

            for row in csv_reader:
                if line_count == 0:
                    line_count = 1
                else:
                    try:
                        if int(row[0]) == state+1:
                            weight = field[str(row[2])]
                            stack[state] = 1 * weight
                        else:
                            if str(row[2]) == label:
                                weight = field[str(row[2])]
                                stack[int(row[0])-1] = 0.1 * weight
                    except KeyError:
                        stack[int(row[0])-1] = 0
                            
        reward_matrix = np.column_stack((reward_matrix, stack))
        state = s_next
        
        if max_epsilon >= min_epsilon:
            decreased_epsilon = decreased_epsilon - (max_epsilon - min_epsilon)/time # epsilon 점차적으로 감소
        else:
            print("Max epsilon should bigger than min epsilon")
            raise RuntimeError
            
        cumulative_rewards[state] += reward
    for ch,value in enumerate(cumulative_rewards):
        if value > threshold:
            channels.append(ch)
            
    return channels


N = 274
data = 3000
time = 1000

max_epsilon = 0.01
min_epsilon = 0.01
eta = 0.1
gamma = 0.1
threshold = 100.0 #5.0

random.seed(1)
np.random.seed(1)
reward_matrix = make_reward_data(N,data,channel)
# 학습 알고리즘 테스트.
results = test_algorithm(reward_matrix, max_epsilon, min_epsilon, eta, gamma, channel, time, threshold)

labels = []
#print(channel)
for i in results:
    if i not in channel:
        results.remove(i)
results.remove(45)

for i in results:
    with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        line_count = 0

        for row in csv_reader:
            if line_count == 0:
                line_count = 1
            else:
                if int(row[0]) == (i+1):
                    labels.append(row[2])
                    
#pprint(results)
#pprint(labels)
for i in range(len(results)):
    results[i] += 1
results = np.array(results)
labels = np.array(labels)
#print(len(results))
#print(len(labels))
final = np.column_stack((results,labels))
pprint(final)
print("Total Reward 임계치:", threshold)
#print("# of channels:", len(results))


# 초기 상태의 TV 모습

# 전체 그림의 크기 및 그림을 나타내는 변수 선언
fig1 = plt.figure(figsize=(17, 5))
ax = plt.gca()

# 장르 표시
plt.text(0.5, 2.5, 'homeshopping', size=14, ha='center')
plt.text(1.5, 2.5, 'movie', size=14, ha='center')
plt.text(2.5, 2.5, 'drama', size=14, ha='center')
plt.text(3.5, 2.5, 'variety', size=14, ha='center')
plt.text(4.5, 2.5, 'music', size=14, ha='center')
plt.text(5.5, 2.5, 'fun', size=14, ha='center')

plt.text(0.5, 1.5, 'sports', size=14, ha='center')
plt.text(1.5, 1.5, 'culture', size=14, ha='center')
plt.text(2.5, 1.5, 'health', size=14, ha='center')
plt.text(3.5, 1.5, 'information', size=14, ha='center')
plt.text(4.5, 1.5, 'adult', size=14, ha='center')
plt.text(5.5, 1.5, 'animation', size=14, ha='center')

plt.text(0.5, 0.5, 'economy', size=14, ha='center')
plt.text(1.5, 0.5, 'refinement', size=14, ha='center')
plt.text(2.5, 0.5, 'religion', size=14, ha='center')
plt.text(3.5, 0.5, 'fishing', size=14, ha='center')
plt.text(4.5, 0.5, 'education', size=14, ha='center')

# 그림을 그릴 범위 및 눈금 제거 설정
ax.set_xlim(0, 6)
ax.set_ylim(0, 3)
plt.tick_params(axis='both', which='both', bottom=False, top=False,
                labelbottom=False, right=False, left=False, labelleft=False)

# 별점 표시
homeshopping = 0
movie = 3
drama = 2
variety = 0
music = 4
fun = 3
sports = 1
culture = 4
health = 0
information = 0
adult = 0
animation = 0
economy = 0
refinement = 0
religion = 0
fishing = 0
education = 0
stars = [homeshopping, movie, drama, variety, music, fun, sports, culture,
         health, information, adult, animation, economy, refinement,
         religion, fishing, education]

for i in range(homeshopping):
    i = i * 0.1
    line, = ax.plot([0.5+i-0.05*(homeshopping-1)], [2.3], marker="*", color='y', markersize=12)
for i in range(movie):
    i = i * 0.1
    line, = ax.plot([1.5+i-0.05*(movie-1)], [2.3], marker="*", color='y', markersize=12)
for i in range(drama):
    i = i * 0.1
    line, = ax.plot([2.5+i-0.05*(drama-1)], [2.3], marker="*", color='y', markersize=12)
for i in range(variety):
    i = i * 0.1
    line, = ax.plot([3.5+i-0.05*(variety-1)], [2.3], marker="*", color='y', markersize=12)
for i in range(music):
    i = i * 0.1
    line, = ax.plot([4.5+i-0.05*(music-1)], [2.3], marker="*", color='y', markersize=12)
for i in range(fun):
    i = i * 0.1
    line, = ax.plot([5.5+i-0.05*(fun-1)], [2.3], marker="*", color='y', markersize=12)

for i in range(sports):
    i = i * 0.1
    line, = ax.plot([0.5+i-0.05*(sports-1)], [1.3], marker="*", color='y', markersize=12)
for i in range(culture):
    i = i * 0.1
    line, = ax.plot([1.5+i-0.05*(culture-1)], [1.3], marker="*", color='y', markersize=12)
for i in range(health):
    i = i * 0.1
    line, = ax.plot([2.5+i-0.05*(health-1)], [1.3], marker="*", color='y', markersize=12)
for i in range(information):
    i = i * 0.1
    line, = ax.plot([3.5+i-0.05*(information-1)], [1.3], marker="*", color='y', markersize=12)
for i in range(adult):
    i = i * 0.1
    line, = ax.plot([4.5+i-0.05*(adult-1)], [1.3], marker="*", color='y', markersize=12)
for i in range(animation):
    i = i * 0.1
    line, = ax.plot([5.5+i-0.05*(animation-1)], [1.3], marker="*", color='y', markersize=12)

for i in range(economy):
    i = i * 0.1
    line, = ax.plot([0.5+i-0.05*(economy-1)], [0.3], marker="*", color='y', markersize=12)
for i in range(refinement):
    i = i * 0.1
    line, = ax.plot([1.5+i-0.05*(refinement-1)], [0.3], marker="*", color='y', markersize=12)
for i in range(religion):
    i = i * 0.1
    line, = ax.plot([2.5+i-0.05*(religion-1)], [0.3], marker="*", color='y', markersize=12)
for i in range(fishing):
    i = i * 0.1
    line, = ax.plot([3.5+i-0.05*(fishing-1)], [0.3], marker="*", color='y', markersize=12)
for i in range(education):
    i = i * 0.1
    line, = ax.plot([4.5+i-0.05*(education-1)], [0.3], marker="*", color='y', markersize=12)
    
    
    
aa = []
ab = []
ac = []
ad = []
ae = []
af = []
ag = []
with open('./folder/dataset.csv', 'r', encoding='UTF-8') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    
    for row in csv_reader:
        if line_count == 0:
            line_count = 1
        else:
            if str(row[2]) == 'movie':
                aa.append(1)
            elif str(row[2]) == 'drama':
                af.append(1)
            elif str(row[2]) == 'culture':
                ac.append(1)
            elif str(row[2]) == 'sports':
                ad.append(1)
            elif str(row[2]) == 'fun':
                ae.append(1)
            elif str(row[2]) == 'music':
                ag.append(1)
print()        
print("# of movie channels:", len(aa))
print("# of drama channels:", len(af))
#print("# of variety channels:", len(ab))
print("# of music channels:", len(ag))
print("# of fun channels:", len(ae))
print("# of sports channels:", len(ad))
print("# of culture channels:", len(ac))
```
